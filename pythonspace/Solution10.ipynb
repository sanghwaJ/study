```python
# Softeer - 장애물 인식 프로그램

import sys
#input = sys.stdin.readline

N = int(input())

blocks = []
visited = []
for i in range(N):
    blocks.append(input())
    visited.append([False] * N)
    
dx, dy = [0, 1, 0, -1], [1, 0, -1, 0]

cnt = 0
for i in range(N):
    for j in range(N):
        if blocks[i][j] == '1' and not visited[i][j]:
            visited[i][j] = True
            cnt += 1
            for w in range(4):
                ii, jj = i + dx[w], j + dy[w]
                if ii < 0 or ii == N or jj < 0 or jj == N:
                    continue
                if blocks[ii][jj] == '1' and not visited[ii][jj]:
                    visited[i][j] = True
                    cnt += 1
                
        
    

```
